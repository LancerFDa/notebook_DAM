## Iterarions and Comprehensions


Cuando pensamos en operaciones que sobre cada item de una secuencia, entramos en las *list comprehensions*.

```python
L = [1, 2, 3, 4, 5]
for i in range(len(L)):
    L[i] += 10
L
# [11, 12, 13, 14, 15]
```

CÃ³digo convencional

---

```python
L = [x + 10 for x in L]
L
# [21, 22, 23, 24, 25]
```

```python
res = []
for x in L:
	res.append(x + 10)
res
# [31, 32, 33, 34, 35]
```
