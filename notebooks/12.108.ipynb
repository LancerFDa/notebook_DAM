# Funcional Programing Tools 

En Python hay distintos paradigmas de programación:

- Procedural
- Orientado a objetos
- Funcional

Para el último, Python incluye herramientas built-ins, que __aplican funciones a sequencias y otros iterables__. Estas son:

- `map`: Herramienta que llama funciones en un item iterable.
- `filter`: Filtra items en basado en una función.
- `reduce`: Aplica funciones a pares de items.

## Anonymous Functions: lamba

Esta expresión crea una función que será llamada luego, devuelve la funcion en vez de asignarle un nombre.

`lambda argument1, argument2,... argumentN : expression using arguments`

- Lambda es una expresión, no una declaración. Por esto, lambda puede aparecer un sitios que `def` no por la sintaxis de Python.
- El cuerpo de lambda es una expresión simple, no un bloque de declaraciones.

```python
def func(x, y, z): 
    return x + y + z
func(2, 3, 4)
# 9
```

```python
f = lambda x, y, z: x + y + z
f(2, 3, 4)
# 9
```

### Why Use lambda?

```python
L = [lambda x: x ** 2,      # Inline function definition
     lambda x: x ** 3,      # A list of three callable functions
     lambda x: x ** 4]

for f in L: print(f(2))

print(L[0](3))
```
