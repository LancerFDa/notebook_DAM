## Basic List Operators

Las listas al ser secuenciales, soportan bastantes de los operadores de los strings.

`len` en una funcion __built-in__.
`len` es un diminutivo de _lenght_, este operador, cuenta la cantidad de valores que tiene una lista.

```python
# Length
len([1, 2, 3])          # 3
```

```python
len(['one', [2, 3], 'four'])    # returns 3, not 4
```

Igual que los _Strings_ las listas tambien se __concatenan__.

Al usar el operador `+` se concatenan, se juntan unas despues de otras.

```python
# Concatenation 
[1, 2, 3] + [4, 5, 6]       # [1, 2, 3, 4, 5, 6]
```

Pero al usar el `*`, tanto los Strings como las listas de duplican esa contidad de veces

```python
# # Repetition
cadena = ['Ni!'] * 4             # ['Ni!', 'Ni!', 'Ni!', 'Ni!']
lista = [1, 2, 3] * 4           # [1, 2, 3, 1, 2, 3, 1, 2, 3, 1, 2, 3]
print (cadena)
print (lista)
```

`str` es una __built-in__, todo lo que esté dentro de esta, pasa a ser String.
`list` es otra __built-in__, todo lo que esté dentro pasa a ser una __lista de carácteres__.

```python

str([1, 2]) + "34"      # Same as "[1, 2]" + "34"
```

En este caso, al usar `str` obligas a que "[1, 2]" sea string y al poner las comillas a __34__ lo haces texto, por lo cual, se concatena en vez de sumarse.

```python
[1, 2] + list("34")     # Same as [1, 2] + ["3", "4"]
```

`[1, 2]` es una lista, y `list` hace que "34" se transforme en una __lista de carácteres__ en vez de ser un Strin, por tanto al sumarlos se concatenan.

### List Iteration and Comprehesion

Una __iteración__ es un porceso que se repite ne cada elemeto.

__value__ `in` es una keyword que usa python para repasar cada elemento de una variable, en busca del valor pedido.
`in` suele ir acompañado de `for`, este es un bucle que se repite hasta que acabe su condición.

```python
# Checking whether an item is in a list
3 in [1, 2, 3]          # True
```

`in` es una keyword que al estar sola, devuelve un valor __ooleano__.
Si es cieta la condición, "True", sino, "False".

```python
for x in [1, 2, 3]:
    print(x, end='')
```

Aqui `in` al esta acompañada de `for` no dan un valor booleano, sino que es un bucle.
En este ejemplo, recoge la lista y la guarda en x, y luego con `print(x, end=" ")` lo imprime con espacios entre los números.

una __List comprehesion__ es una forma de crear uan nueva lista aplicando una expresión a cada elemento de la secuencia.º

```python
res = [c * 4 for c in 'SPAM']
res
```

```python
#CRUD
res = []                # Create
for c in 'SPAM':        # Read
    res.append(c * 4)   # Update
del res[0]              # Delete
res
```

```python
list(map(abs, [-1, -2, 0, 1, 2]))      # Map a function across a sequence [1, 2, 0, 1, 2]
```

### Indexing, Slicing and Matrixes

El __Index__ o __índice__ de un elemto es su posición, es decir __Indexing__.

__Indexing__ o __indexado__ es el proceso de acceder a un elemento usando su posición dentro de una *cadena*, *lista* o *tupla*.

```python
L = ['spam', 'Spam', 'SPAM!']
L[2]   
```

Las listas empiezan en el __0__, por eso al poner `L[2]` da el tercer elemento.

```python
L[-2]               # 'Spam'
```

Poniendolo negativo, cuenta desde izquierda a derecha.

```python
L[1:]               # ['Spam', 'SPAM!']
```

Con los `[:]` cuentas a partir de los dos puntos, si está a la derecha del número, cuentas los elementos que estén en ese lado, si esta a la izquierda, no cuenta los que están a la derecha de dicho elemento.

```python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
matrix[1]
```

```python
matrix[2][0]
```

```python
matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]
matrix[1][1]
```

con `matrix` es algo similar, se divide en listas (0, 1, 2...) si pones `[1][1]`, te refieres a la segunda tabla y su segundo elemento.

## Changing Lists in Place

Como las listas son __mutables__, se pueden cambiar elementos en su lugar.
Esto te permite modificar la lista sin necesidad de hacer una copia.

### Index and slice assigments

El __slice assigment__ es modificar una lista "cortandola" y sustituyendo esos valores con otros nuevos.

```python
L = ['spam', 'Spam', 'SPAM!']
L[1] = 'eggs'                       # Index assignment
L                                   # ['spam', 'eggs', 'SPAM!']
```

sustituyes `spam` por `egg`.

```python
L[0:2] = ['eat', 'more']            # Slice assignment: delete+insert, replaces items 0,1
L                                 # ['eat', 'more', 'SPAM!']
```

sustituyes los dos primeros valores `spam, Spam`[0:2] por `eat, more`

```python
L = [1, 2, 3]                       # Replacement/insertion
L[1:2] = [4, 5]                     # [1, 4, 5, 3]
L
```

`[1:2]` selecciona solo el elemento 2 (selecciona entre después del 1 y antes del 2) y lo remplaza por `[4, 5]`.

```python
L[1:1] = [6, 7]                     # Insertion (replace nothing)
L                                   # [1, 6, 7, 4, 5, 3]
```

Se selecciona `[1:1]` es decir entre 1 y 1, nada, y se añade después del elemento 1 `[6, 7]`.

```python
L[1:2] = []                         # Deletion (insert nothing)
L                                   # [1, 7, 4, 5, 3]
```

Este es al revés, seleccionas el elemento 2 y lo sustituyes por nada, `[]`.

```python
L = [1]
L[:0] = [2, 3, 4]                   # Insert all at :0, an empty slice at front
L
```

Selecciona el espacio que hay detrás del elemento y lo sustituye por `[2, 3, 4]`.

```python
L[len(L):] = [5, 6, 7]              #  Insert all at len(L):, an empty slice at end
L
```

copia `L`con `len(L):` e inserta en la copia `[5, 6, 7]`.

### List method calls

Como en los `strings` las listas tambien tienen un tipo especifico de metodos para modificarlos.

Los `metodos` son clases, solo que aplican a un objeto en particular.

##### apend

```python
L = ['eat', 'more', 'SPAM!']        
L.append('please')                  # Append method call: add item at end
L                                   # ['eat', 'more', 'SPAM!', 'please']
```

`.apend` añade un elemento más a la lista.

Si intentas añadir una lista a otra, la segunda se unira como un solo elemento al final de la otra.

```python
a = [1, 2, 3, 4, 5, 6, 7, 7]
b = [8, 9]
a.append(b)                     # a: [1, 2, 3, 4, 5, 6, 7, 7, [8, 9]]
print(a)
print(a[8])
```

##### sort


```python
L.sort()                            # Sort list items ('S' < 'e')
L                                   # ['SPAM!', 'eat', 'more', 'please']
```

`sort` ordena en orden descenedente.

```python
L = ['abc', 'ABD', 'aBe']           # Sort with mixed case
L.sort()
L                                   # ['ABD', 'aBe', 'abc']
```

```python
L = ['abc', 'ABD', 'aBe']           # L ha sido cambiada in place en la operación sort() anterior, objeto mutable!
L.sort(key=str.lower)
L                                   # ['abc', 'ABD', 'aBe']
```

##### extend

extiende la lista añadiendo elementos de otros enumerables.

```python
L = [1, 2]
L.extend([3, 4, 5])                 # [1, 2, 3, 4, 5]
L.pop()                             # Delete and return last item (by default: −1): 5
L
```

```python
a = [1, 2, 3, 4, 5, 6, 7, 7]
b = [8, 9, 10]
a.extend(b)                         # a: [1, 2, 3, 4, 5, 6, 7, 7, 8, 9, 10]
a.extend(range(3))                  # Extend list with elements from a non-list enumerable:
a                                   # a: [1, 2, 3, 4, 5, 6, 7, 7, 8, 9, 10, 0, 1, 2]
```

#### Any and All

`all` determina si todos los valores son `True`.

```python
nums = [1, [1], 'a', 1]
all(nums)
# True
```

```python
chars = ['a', 'b', 'c', 'd']
all(chars)
# True
```

`any` determina si uno o más son `True`.

```python
nums = [1, 1, 0, 1]
any(nums)
# True
```

```python
vals = [None, 0, [], False, '']
any(vals)
# False
```
