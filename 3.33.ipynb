## OBJETOS MUTABLES

las `listas` son mutables, "a" tiene de valos `2, 3, 4` y "b" tiene como valor "a", si cambias el valor de "a" también varía "b".
En memoria como usan el mismo valor, varía su contenido.

```python
L = [2, 3, 4]	    # A mutable object
M = L			    # Make a reference to the same object
L[0] = 24 		    # An in-place change

print("L = ", L)    # [24, 3, 4]
print("M = ", M)    # [24, 3, 4]
```

```python
L = [2, 3, 4]
M = L[:]		    # Make a copy of L (or list(L), copy.copy(L), etc.)
L[0] = 24

print("L = ", L)    # [24, 3, 4]

print("M = ", M)    # [2, 3, 4]  M is not changed

L =  [24, 3, 4]
```

Con `[:]` hacemos una copia del valor de "L" en "M", en distintos espacios de memoria, al modificar el primer valor de "L" con `L = [0] = 24` "L" y "M" tienen lista y valores distintas.


---

En estos ejemplos se usa `==` y `is`.
`==` se usa para comprobar que ambas variables tienen el mismo valor.
`is` es para camprobar que las dos variables ocupan el mismo espacio de memoria.

```python
L = [1, 2, 3]
M = L 			    # M and L reference the same object

L == M 			    # True: Same values => operador igualdad de valores
L is M 			    # True: operador identidad de objetos: compara las referencias (los punteros)
```

Como `M = L`, tienen el mismo valor `L == M --> True`, también usan el mismo espacio de memoria `L is M --> True`.

```python
L = [1, 2, 3]
M = [1, 2, 3]	    # M and L reference different objects

print(
    "L == M",
    L == M 			# True: Same values
)

print(
    "L is M", 	    # False: operador identidad de objetos: compara las referencias (los punteros)
    L is M
)
```

Como `L = [1, 2, 3]` y `L = [1, 2, 3]`, tienen el mismo valor `L == M --> True`, pero no usan el mismo espacio de memoria `L is M --> False`.

## CACHE


```python
X = 42
Y = 42		    # Should be two different objects

X  == Y         # True
X is Y 		    # True: Same object anyhow: caching at work!!
```

Cuando es un `integer` al definir "X" e "Y" con la misma integer, usan el mismo sitio de memoria.

```python
import sys
sys.getrefcount(1)		# number of pointers to this shared piece of memory
```
